### Cleaning the Columns for Future Analysis (Revise Title)

```python
cols_for_booleans = ['Have you seen any of the 6 films in the Star Wars franchise?',
                     'Do you consider yourself to be a fan of the Star Wars film franchise?']

# Gauging the types of boolean-like values

for col in cols_for_booleans:
    
    print(star_wars[col].value_counts(dropna=False),'\n')

# The column may be an object type, but the boolean pattern is 
# very prevalent. 
```

```python
vals_for_bools = {'Yes':True,'No':False}

# Mapping to booleans for future analyses

for col in cols_for_booleans:
    
    star_wars[col] = star_wars[col].map(vals_for_bools)
    
    print(star_wars[col].value_counts(dropna=False),'\n')
```

These columns had many yes/no, boolean-like answers, so they were mapped appropriately. The next step is to clean the columns relating to the films themselves, which also has a somewhat boolean format as well. Value counts of them will be displayed.

```python
for col in star_wars.columns[3:9]:
    
    print(star_wars[col].value_counts(dropna=False),'\n')
    
    # The questions about viewership of said films, key columns
```

The first question corresponds to the first movie in the hexalogy. The "Unnamed: 4" corresponds to the second movie, and "Unnamed: 5" matches the third movie, and so on and so forth. If the person saw the movie, the answer is the movie title; in contrast, a null value corresponds to either a respondent who did not answer or did not watch the movie. For simplicity's sake, the latter will be assumed. The next move is to replace those column names and replace the values with booleans.

```python
hexalogy_dict = {} 

counter = 1 
# counter contains a starting value and increases to match all the 
# numbered entries of the hexalogy's order.

for col in star_wars.columns[3:9]:
    
    hexalogy_dict[col] = 'seen_{}'.format(counter)
    
    counter+=1
    
print(hexalogy_dict)
    
    
```

```python
star_wars.rename(columns = hexalogy_dict, inplace=True) 

star_wars.columns[:9]

# The viewership columns have been successfully fixed.
```

```python
# Converting the movie column values to booleans

movie_cols = [col for col in star_wars.columns if 'seen_' in col]

star_wars[movie_cols] = (star_wars[movie_cols].applymap(
    lambda val : False if pd.isnull(val) else True)
                        )


star_wars.head(3)
    
```

### Film Ratings and Viewership by Subgroups

Since the target subgroups focus on gender, fandom, and wages, writing a function that can speed up a tedious and repetitive process is a logical solution. Said solution's approach begins with two subplots, one for ratings and another for viewership. Within those, bars will be plotted based on the series data from two groups. For example, gender subdivides itself into male and female for this study. Lastly, the ratings and viewership data will use means and sums respectively. 

```python
def rank_and_viewership (category):
    
    categories = {
        'gender':['red','green'],
        'star_wars_fan' : ['blue','orange'],
        'star_trek_fan' : ['brown','gray']
                 }
    
    # These categories determine the type of study.
    # They also hold the colors for group 1 and group 2.
    
    
    if category not in categories:

        category_list = [x for x in categories.keys()]

        return print('Usable categories : ', category_list)
        
        # This is an equivalent error message if someone
        # were to use this function for this type of study.

    
    # Defining the entirety of the subgroups 
    
    if category == 'gender':
        
        df1 = star_wars[star_wars["Gender"] == "Male"] # group 1
        
        df2 = star_wars[star_wars["Gender"] == "Female"] # group 2
        
        label_1 = 'Male'
        
        label_2 = 'Female'
        
        rank_graph_title = 'Movie Rankings by Gender' 
        
        viewer_graph_title = 'Respondent Viewership by Gender'
         
            
    elif category == 'star_wars_fan':
        
        col_string = 'Do you consider yourself to be a fan of the Star Wars film franchise?'
        
        df1 = star_wars[star_wars[col_string]== True]
        
        df2 = star_wars[star_wars[col_string]== False]
        
        label_1 = 'Fan'
        
        label_2 = 'Non-Fan'
        
        rank_graph_title = 'Movie Rankings by SW Fans Vs. Non-Fans'
        
        viewer_graph_title = 'Respondent Viewership by SW Fans Vs. Non-Fans'
        
        
    elif category == 'star_trek_fan':
        
        col_string = 'Do you consider yourself to be a fan of the Star Trek franchise?'
        
        df1 = star_wars[star_wars[col_string]=='Yes']
        
        df2 = star_wars[star_wars[col_string]=='No']
        
        label_1 = 'Fan'
        
        label_2 = 'Non-Fan'
        
        rank_graph_title = 'Movie Rankings by ST Fans Vs. Non-Fans'
        
        viewer_graph_title = 'Respondent Viewership by ST Fans Vs. Non-Fans'
        
        # Did not include these 3 categories' parameters in the 
        # 'categories' dictionary to have better, readable code.
        
    # Set width of bar and preparing subplots
    bar_width = 0.4
    fig = plt.figure(figsize = (12,12))
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1,2,2)

    # Set position of bars on X axis
    br1 = np.arange(6) # n = 6 for all subplots
    br2 = [x + bar_width for x in br1] 
    tick_positions = range(0,6)

    # the first subplot : rankings

    ranking_means_1 = df1.iloc[:,9:15].mean()

    ranking_means_2 = df2.iloc[:,9:15].mean()

    ranking_cols = star_wars.columns[9:15]

    ax1.bar(br1, ranking_means_1, color = categories[category][0], 
            width = bar_width, edgecolor ='grey', label = label_1) 

    ax1.bar(br2, ranking_means_2, color =categories[category][1], 
            width = bar_width, edgecolor ='grey', label = label_2) 

    ax1.set_xticks(tick_positions)

    ax1.set_xticklabels(ranking_cols, rotation = 'vertical')

    ax1.set_ylim(0,5) # Average rating values don't exceed 5.

    ax1.set_ylabel('Rank (lower is better)', fontsize = 13)

    ax1.set_title(rank_graph_title, 
                  fontweight = 'bold', fontsize = 13)


    # the second subplot : respodent viewership per movie by groups

    seen_sums_1 = df1.iloc[:,3:9].sum()

    seen_sums_2 = df2.iloc[:,3:9].sum()

    seen_cols = star_wars.columns[3:9]

    ax2.bar(br1, seen_sums_1, color = categories[category][0], 
            width = bar_width, edgecolor ='grey', label = label_1) 

    ax2.bar(br2, seen_sums_2, color = categories[category][1], 
            width = bar_width, edgecolor ='grey', label = label_2) 

    ax2.set_xticks(tick_positions)

    ax2.set_xticklabels(seen_cols, rotation = 'vertical') 

    ax2.set_ylim(0,600) # 600 provides an adequate zoom on the bars.

    ax2.set_ylabel('Viewership Count', labelpad = 4.0, fontsize = 13)

    ax2.set_title(viewer_graph_title, 
                  fontweight = 'bold', fontsize = 13) 


    # Displaying the plot

    plt.legend()

    plt.show()
```

### Socioeconomic Status

Before plotting anything from the perspective of socioeconomic status, it is important to understand the types of values that populate the column.

```python
star_wars['Household Income'].value_counts(dropna=False)
```

The approach to preparing the data is twofold. The salary names have to be cleaned for better readability because they will be used as labels. The other part of the process is to prepare two groupings of the dataframe, one with means and one with sums.

```python
# Abbreviating the salary ranges

def fix_salary(val):
    
    import re
    
    if pd.isnull(val):
        
        return np.nan # Nans will not be represented as another value
        
    pat = r",\d+"   
    
    fixed_string = re.sub(pat,'K',val) #30,000 to 30K for example
    
    if '$' in fixed_string:
        
        return fixed_string.replace('$','')
    
    return fixed_string


star_wars['Household Income'] = star_wars['Household Income'].apply(fix_salary)


# Aggregating via means for ratings and sums for viewership

wage_groups_means = star_wars.groupby('Household Income').agg(np.mean)

wage_groups_means.reset_index(inplace=True)

wage_groups_sums = star_wars.groupby('Household Income').agg(np.sum)

wage_groups_sums.reset_index(inplace=True)

# Index was reset to use wage brackets as booleans and labels

print('Wage Group Categories for Means : ', '\n',
      wage_groups_means.iloc[0:,0],'\n')

print(wage_groups_means.iloc[:,9:15],'\n',)

print('Wage Group Categories for Sums : ', '\n',
      wage_groups_sums.iloc[0:,0],'\n')

print(wage_groups_sums.iloc[:,3:9],'\n')

# The order of the categories is the same across both aggregations.
# They will be organized for plotting in the next block.
```

```python
cat_order = [0,3,4,1,2] # To have salaries from lowest to highest
                        # 0-24K, 25K-49K, 50K-99K, 100K-149K, 150K+

# The wage labels are the same for both sets.
wage_cat = [wage_groups_means.iloc[cat,0] for cat in cat_order]
    
# Set width of bar 
bar_width = 0.15 
fig_x, (ax_A, ax_B) = plt.subplots(2, 1, figsize = (12,10))

# Setting up the bars for the loops
bar_num = np.arange(len(ranking_cols)) # n = 6 for all subplots
tick_positions = range(0,6) 
color_combo = {
    0 : '#f3b780',
    1 : '#c4e089', 
    2 : '#89d9e0' ,
    3 : '#8997e0' ,
    4 : '#d389e0',
    5 : '#da9595'
} # Color for the wage groups


# The label names for the subplots

ranking_cols = star_wars.columns[9:15] # Ratings

seen_cols = star_wars.columns[3:9] # Viewership


for num in np.arange(5): 
    
    category = wage_cat[num] 
    
    # Series for that wage bracket, 
    # one for rank and one for viewership
    
    wage_group_rank = (
        wage_groups_means[wage_groups_means.iloc[:,0] == category]
        .iloc[0,9:15])
    
    wage_group_view = (
        wage_groups_sums[wage_groups_sums.iloc[:,0] == category]
        .iloc[0,3:9])
    
    bar_loop = [x + bar_width*(num) for x in bar_num] 
    
    # 5 bars per bracket, so the bar_width has to adjust accordingly.
        
    
    ax_A.bar(bar_loop, wage_group_rank, color = color_combo[num], 
            width = bar_width, edgecolor ='grey', label = category) 

    ax_B.bar(bar_loop, wage_group_view, color = color_combo[num], 
            width = bar_width, edgecolor ='grey', label = category) 

    
ax_A.set_xticks(tick_positions)

ax_A.set_xticklabels(ranking_cols)

ax_A.set_ylim(0,5) # Ratings do not exceed 5.

ax_A.set_ylabel('Rank (lower is better)', fontsize = 13)

ax_A.set_title('Movie Ranking by Wage Group', 
              fontweight = 'bold', fontsize = 14)


ax_B.set_xticks(tick_positions)

ax_B.set_xticklabels(seen_cols) 

ax_B.set_ylim(0,350) # 350 serves to better zoom on the image.

ax_B.set_ylabel('Viewership Count', labelpad = 4.0, fontsize = 13)

ax_B.set_title('Respondent Viewership by Wage Group', 
              fontweight = 'bold', fontsize = 14) 


#Displaying the plot

plt.legend()

plt.show()
```
